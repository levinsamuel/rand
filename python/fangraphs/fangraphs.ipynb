```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
fgdf = pd.read_csv('FanGraphsLeaderboard.csv').set_index('playerid')
fgdf['BB%'] = fgdf['BB%'].apply(lambda p: float(p[:-1]))
fgdf['K%'] = fgdf['K%'].apply(lambda p: float(p[:-1]))

fgdf.head(20)
```

```python
def label_subset(df, name_list):
    ja = fgdf[fgdf['Name'].isin(name_list)]
    noja = fgdf[~fgdf['Name'].isin(name_list)]
    return ja, noja

ja, noja = label_subset(fgdf, ['Jorge Alfaro', 'Joey Votto'])
ja
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6,4))

ax.scatter(noja['K%'], noja['BB%'], alpha=.5)

for i, row in ja.iterrows():
#     print(s)
    ax.scatter(row['K%'], row['BB%'], alpha=.5)

ax.set_xlabel("K percentage")
ax.set_ylabel("BB percentage")

```
